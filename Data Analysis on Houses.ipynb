# **Shapiro-Wilk test**
Tests whether a data sample has a Gaussian distribution.

Null Hypothesis (H0): The sample adheres to a Gaussian distribution.

Alternative Hypothesis (H1): The sample deviates from a Gaussian distribution.






```python
from scipy.stats import shapiro

def shapiro_test(data):
    statistic, p_value = shapiro(data)
    print(f'Statistic: {statistic:.3f}')
    print(f'p-value: {p_value:.3f}')
```

```python
cols_ = df['LotArea']
shapiro_test(cols_)
```

Because our p-value is below 0.05, we can totally declare that 'LotArea' ain't playing by the normal distribution rules.

# **Anderson-Darling test**
We've got a bunch of tests to see if our data is playing it cool with that normal distribution vibe. Right now, we're rolling with the Anderson-Darling test on 'overallQual'

```python
from scipy.stats import anderson
temp = anderson(df['OverallQual'])
print(f"Test statistic: {temp.statistic:.3f}")
print(f"Critical values: {temp.critical_values}")
```

```python
import matplotlib.pyplot as plt
plt.hist(df['OverallQual'], bins=15 , color = 'white' , edgecolor = 'red')
plt.title('Overall Quality')
plt.show()
```

As you can see the test statistic is larger than any of the critical values, so the null hypothesis is rejected and  this feature ain't playing nice with a normal distribution.

# **Analysis of Variance test**

This test only works for numerical features; for categorical ones, we must try other methods. Here, we group our target column by the neighborhood's different values and then apply the ANOVA test.

This test has two hypotheses:

Null Hypothesis (H0): No statistically significant distinction exists in the means of the features under comparison.

Alternative Hypothesis (H1): A statistically significant difference exists between the means.

```python
ANOVA_df = df.copy(deep=True)
```

```python
from scipy import stats
anova = ANOVA_df.groupby('Neighborhood')['SalePrice']
f_value , p_value = stats.f_oneway(*[group for name , group in anova])
print('f={0:.3f} , p={1:.3f}'.format(f_value,p_value))
```

There is a significant difference between the means of two features,these two features are correlated.

```python
anova = ANOVA_df.groupby('SaleCondition')['SalePrice']
f_value , p_value = stats.f_oneway(*[group for name , group in anova])
print('f={0:.3f} , p={1:.3f}'.format(f_value,p_value))
```

Sale condition is related to sale price however of lower f-value,the relationship between them is less strong compared to Neighborhoof and Sale price.

# **T-test**

We use t-test to figure out if two groups of data are different from each other or not and if they are , how significant this difference is.


```python
from scipy.stats import ttest_ind
test_df = df.copy(deep=True)
t_statistic, p_value = ttest_ind(test_df['SalePrice'], test_df['2ndFlrSF'])
print('T-Statistic:', t_statistic)
print('P-Value:', p_value)
```

This result makes us certain that there's a large difference between the means of the two columns we chose. Since the p-value is 0.0 we can conclude that this difference hasn't happened by chance.
